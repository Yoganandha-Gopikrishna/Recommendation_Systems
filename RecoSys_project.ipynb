## Q3 - Build a popularity based model and recommend top 5 mobile phones.

```python
ppt_data_grouped = final_df.groupby('product').agg({'score': 'count'}).reset_index()
ppt_data_grouped.head()
```

```python
#Sort the products on recommendation score 
ppt_data_sort = ppt_data_grouped.sort_values(['score', 'product'], ascending = [0,1]) 
      
#Generate a recommendation rank based upon score 
ppt_data_sort['Rank'] = ppt_data_sort['score'].rank(ascending=0, method='first') 
          
#Get the top 5 recommendations 
popularity_recommendations = ppt_data_sort.head(5) 
popularity_recommendations 
```

## Q4 - Build a collaborative filtering model using SVD. You can use SVD from surprise or build it from scratch.

## Q5 - Evaluate the collaborative model. Print RMSE value.

```python
import surprise
from surprise import Dataset,Reader
```

```python
reader = Reader(rating_scale=(1, 10))
final_data = Dataset.load_from_df(final_df[[ 'author', 'product','score']], reader)
```

```python
# Split data to train and test

from surprise.model_selection import train_test_split
trainset, testset = train_test_split(final_data, test_size=.20,random_state=612)
```

```python
trainset.all_ratings()
```

```python
print(trainset.to_raw_uid(6))
```

```python
print(testset[0])
```

```python
from surprise import SVD, KNNWithMeans
from surprise import accuracy
```

```python
svd_model = SVD(n_factors=5,biased=False,random_state=612)
svd_model.fit(trainset)
```

```python
test_pred = svd_model.test(testset)
```

```python
# get RMSE
print("SVD Model : Test Set")
print(accuracy.rmse(test_pred))
```

### Userbased and Item based collabarative filtering

#### KNNWithMeans

#### user based

```python
algo_user = KNNWithMeans(k=70, sim_options={ 'user_based': True})

algo_user.fit(trainset)

test_pred_u=algo_user.test(testset)
```

```python
# get RMSE
print("User-based Model : Test Set")
accuracy.rmse(test_pred_u, verbose=True)
```

#### Item based

```python
algo_item = KNNWithMeans(k=70, sim_options={ 'user_based': False})

algo_item.fit(trainset)

test_pred_item=algo_item.test(testset)
```

```python
# get RMSE
print("Item-based Model : Test Set")
print(accuracy.rmse(test_pred_item,verbose=True))
```

- By comparing the all the RMSE scores item-item collabarative filtering gives lower value
